# COMBINE DATA

```python
cdcdat['CC'] = cc['percent']
cdcdat['COUGH'] = cough['percent']
cdcdat['FLU_SYMP'] = flu_symp['percent']
cdcdat['FLU'] = flu['percent']
cdcdat['GRIPE'] = gripe['percent']
cdcdat['INF_SYMP'] = inf_symp['percent']
cdcdat['INFL'] = infl['percent']
cdcdat['OSELT'] = oselt['percent']
cdcdat['PREV'] = prev['percent']
cdcdat['VACC'] = vacc['percent']

x_cols = ['CC', 'COUGH', 'FLU_SYMP', 'FLU', 'GRIPE', 'INF_SYMP','INFL', 'OSELT', 'PREV', 'VACC']

for column in x_cols:
    cdcdat[column] = cdcdat[column].replace(to_replace='<1', value=0)
    cdcdat[column] = cdcdat[column].astype(float)

cdcdat['SUM'] = cdcdat['CC'] + cdcdat['COUGH'] + cdcdat['FLU_SYMP'] + cdcdat['FLU'] + cdcdat['GRIPE'] + cdcdat['INF_SYMP'] + cdcdat['INFL'] + cdcdat['OSELT'] + cdcdat['PREV'] + cdcdat['VACC']
cdcdat['AVG'] = cdcdat['SUM']/10

cdcdat['SEASON'] = (cdcdat['WEEK'] >= 40) | (cdcdat['WEEK'] <= 16)
cdcdat['SEASON'] = cdcdat['SEASON'].replace(to_replace=True, value=1)
cdcdat['SEASON'] = cdcdat['SEASON'].replace(to_replace=False, value=0)

cdcdat = cdcdat.drop(columns = ['PERCENT A', 'PERCENT B', 'TOTAL A', 'TOTAL B'])

# cdcdat['SEASON'] 

cdcdat.head(52)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import  MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras.backend as K
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
from keras.models import model_from_json
```

```python
def rmse (y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred -y_true), axis=-1))
```

```python
#inputFeature
# X=['CC','COUGH','FLU_SYMP','FLU','GRIPE','INF_SYMP','INFL','OSELT','PREV','VACC']
# X=['CC','COUGH','FLU_SYMP','GRIPE','INF_SYMP','INFL','OSELT','VACC']

# X=['FLU_SYMP','FLU', 'GRIPE', 'INF_SYMP','INFL','OSELT','VACC', 'WEEK']
X=['COUGH','FLU','GRIPE','INF_SYMP','INFL','OSELT','VACC', 'WEEK', 'SEASON']

# X = ['SEASON', 'INF_SYMP', 'WEEK']
# X = ['GRIPE']
#TargetValue
Xi = cdcdat[X]
y = ['PERCENT POSITIVE']
yi = cdcdat[y]

# X_train, X_val, y_train, y_val = train_test_split(Xi, yi , test_size=0.3)
```

```python
X_train = Xi.loc[0:468]
y_train = yi.loc[0:468]
X_val = Xi.loc[469:625]
y_val = yi.loc[469:625]
```

```python
from keras import Sequential
from keras.layers import Dense
def build_regressor():
    regressor = Sequential()
    regressor.add(Dense(units=1, input_dim=9))
    regressor.add(Dense(64, kernel_initializer='normal', activation='tanh'))
    regressor.add(Dense(64, kernel_initializer='normal', activation='relu'))
    regressor.add(Dense(20, kernel_initializer='normal', activation='relu'))
    regressor.add(Dense(units=1))
    regressor.compile(optimizer='adam', loss='mean_squared_error',  metrics=['mae',rmse])
    return regressor
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
regressor = KerasRegressor(build_fn=build_regressor, batch_size=32,epochs=100)
```

```python
filepath="weights.best2.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_rmse', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
results=regressor.fit(X_train,y_train, validation_data=(X_val, y_val), callbacks=callbacks_list)
# keras.callbacks.ModelCheckpoint(filepath, monitor='val_rmse', verbose=0, save_best_only=True, save_weights_only=False, mode='auto', period=1)
# y_pred= regressor.predict(X_val)
```

```python
# model = model_from_json(open('weights.best2.hdf5').read())
dependencies = {
    'rmse': rmse
}

model = load_model('weights.best2.hdf5', custom_objects=dependencies)
```
