# Shift By 2 Weeks

```python
cases = data[['YEAR','WEEK','Total Cases']]
st = data[['index','influenza st pop','influenza dis pop','cc dis pop','flu st pop','flu symp st pop','influenza vaccine vacc pop','prevent flu st pop']]
i = st.shape[0]
st=st.drop([i-1,i-2])
cases=cases.drop([0,1])
st=st.reset_index()
cases=cases.reset_index()
cases=cases.reset_index()
# yes, I deliberately reset cases' index twice
cases.head()
shifted_data_2 = pd.concat([cases,st],axis=1)
shifted_data_2=shifted_data_2.drop(columns='level_0')
shifted_data_2.to_csv("2020shiftedByTwoWeeks.csv")
```
