```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import time
from sklearn import preprocessing
import matplotlib
```

# Trends Data Graph

```python
data = pd.read_csv('fulldataset2020.csv')
```

```python
data.head()
```

```python
data = data.drop(columns="Unnamed: 0")
```

```python
matplotlib.rcParams.update({'font.size': 20})
fig = plt.figure(figsize=(20,10))
x = np.arange(len(data))
y = data['cc dis pop']
plt.plot(x,y)
plt.suptitle('Common Cold Over Time')
plt.xlabel('Number of Weeks Starting at March 13, 2005')
plt.ylabel('Popularity of \'Common Cold\' Disease Search Term')
plt.show()
```
